## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/jjmaturino/bloodhound.git
```

or from [conda][conda]

```sh
$ conda install -c jjmaturino bloodhound
```

or from [pypi][pypi]


